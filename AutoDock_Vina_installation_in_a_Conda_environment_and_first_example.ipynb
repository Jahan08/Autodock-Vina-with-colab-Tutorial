## To split the poses

```python
def split_poses(file):

        poses = []

        temp_data = []



        with open(file, 'r') as f:

            lines = f.readlines()

            for line in lines:

                if line.startswith('ENDMDL'):

                    if len(temp_data) > 0:

                        poses.append(temp_data)

                        temp_data=[]

                elif 'HETATM' in line or 'ATOM' in line:

                    temp_data.append(line)



        output = file.split('.')[0]
```
